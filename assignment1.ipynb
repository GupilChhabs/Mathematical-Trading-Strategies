```python
import yfinance as yf
import statistics as stat
```

```python
apple=yf.Ticker("aapl")
apple_history=apple.history(start="2010-01-01",end="2023-05-01",interval="1d")
apple_history["High"]
```

```python
apple_highest=max(apple_history["High"])
```

```python
print(apple_highest)
```

```python
apple_lowest=min(apple_history["Low"])
apple_mdd=(apple_lowest-apple_highest)*100/apple_highest
print(apple_mdd)
```

```python
apple_history
```

```python
apple_history["Open"][0]
```

```python
apple_CR=(apple_history["Open"][-1]-apple_history["Open"][0])*100/apple_history["Open"][0]
print(apple_CR)
```

```python
apple_history["return"]=apple_history["Close"]-apple_history["Open"]
print(apple_history["return"])
```

```python
apple_avgreturn=stat.mean(apple_history["return"])
print(apple_avgreturn)
```

```python
apple_riskfree=((apple_history["Open"][0])*(1+5.25/100)**(4869/365)-apple_history["Open"][0])/4869
```

```python
print(apple_riskfree)
```

```python
apple_std=stat.stdev(apple_history["return"])
print(apple_std)
```

```python
apple_sharpe=(apple_avgreturn-apple_riskfree)*(252**0.5)/apple_std
```

```python
print(apple_sharpe)
```

```python
a=[]
i=0
while(apple_history["return"][i]):
    if(apple_history["return"][i]<0):
        a.append(apple_history["return"][i])
    i+=1   
apple_downdev=stat.stdev(a)
print(apple_downdev)
```

```python
apple_sortino=(apple_avgreturn-apple_riskfree)*(252**0.5)/apple_downdev
print(apple_sortino)
```

```python
apple_volatility=stat.stdev(apple_history["return"]*(252**0.5))
print(apple_volatility)
```

```python
gspc=yf.Ticker('^N225')
gspc_history=gspc.history(start='2010-01-01',end='2023-05-01',interval='1d')
gspc_history
```

```python
indices=['^GSPC','^DJI','^N100','^FTSE','^N225']
for x in indices:
    index=yf.Ticker(x)
    index_history=index.history(start='2010-01-01',end='2023-05-01',interval='1d')
    index_highest=max(index_history['High'])
    index_lowest=min(index_history['Low'])
    index_mdd=(index_highest-index_lowest)*100/index_highest
    print(index_mdd)

    index_CR=(index_history["Open"][-1]-index_history["Open"][0])*100/index_history["Open"][0]
    print(index_CR)

    index_history["return"]=index_history["Close"]-index_history["Open"]
    index_avgreturn=stat.mean(index_history["return"])
    index_riskfree=((index_history["Open"][0])*(1+5.25/100)**(4869/365)-index_history["Open"][0])/4869
    index_std=stat.stdev(index_history["return"])
    index_sharpe=(index_avgreturn-index_riskfree)*(252**0.5)/index_std
    print(index_sharpe)

    a=[]
    i=0
    while(index_history["return"][i]):
        if(index_history["return"][i]<0):
            a.append(index_history["return"][i])
        i+=1
    
    index_downdev=stat.stdev(a)

    index_sortino=(index_avgreturn-index_riskfree)*(252**0.5)/index_downdev
    print(index_sortino)

    index_volatility=stat.stdev(index_history["return"]*(252**0.5))
    print(index_volatility)
    print('\n')
```
